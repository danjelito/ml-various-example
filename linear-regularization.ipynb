```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.datasets import make_regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x, y = make_regression(n_samples=100, n_features=1, noise=10)
poly = PolynomialFeatures(degree=3, include_bias=False)
x_poly = poly.fit_transform(x)
x_poly.shape, y.shape
```

```python
plt.scatter(x_poly[:, 2], y)
plt.show()
```

```python
linear = LinearRegression()
ridge = Ridge(alpha=10)
lasso = Lasso(alpha=10)
elasticnet = ElasticNet()

x = x_poly[:, 2].reshape(-1, 1)

linear.fit(x, y)
ridge.fit(x, y)
lasso.fit(x, y)
elasticnet.fit(x, y)

y_linear = linear.predict(x)
y_ridge = ridge.predict(x)
y_lasso = lasso.predict(x)
y_elasticnet = elasticnet.predict(x)

plt.scatter(x, y)
plt.plot(x, y_linear, label="linear")
plt.plot(x, y_ridge, label="ridge")
plt.plot(x, y_lasso, label="lasso")
plt.plot(x, y_elasticnet, label="elasticnet")
plt.legend()
plt.show()
```
